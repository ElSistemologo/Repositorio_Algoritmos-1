Ejemplo Bioinformático de busqueda por metodo de fuerza bruta


```python
#Ejemplo_profesor
target='CGTTTTCGCTTTTCCGTTTACGATATAT'
a=secuence.find(target)
print(a)
```

```python
#Resolución grupo / Fuerza Bruta
target='CGTTTTCGCTTTTCCGTTTACGATATAT'       #O(1), es la sección de secuencia que buscamos.
prueba=''                                   #O(1), en esta variable se guardan los caracteres a probar.
for i in range(len(secuence)):              #O(O(n)*O(n)), con el for se recorre la secuencia por el índice.
    if secuence[i] == target[len(prueba)]:  #O(n), aqui se compara el target con la secuencia usando "if".
        prueba += secuence[i]               #O(1), en caso de ser iguales, se añade a "prueba"
    else:                                   #O(1)
        prueba=secuence[i]                  #O(1), en caso de no ser iguales, se resetea "prueba"
    if target==prueba:                      #O(n), con esto se revisa cuando se encuentra el target. Esto sucede las n veces que se itere.
        print("Target encontrado desde la posición {} hasta la {}".format((i-len(target)+1),i)) #O(1), en este caso se usa i que es la posición para determinar inicio y final del target en la secuencia.
        break                               #O(1),rompemos el for, por lo que el peor caso es que esto nunca ocurra, al revisar los n elementos.
#O = O(1)+O(1)+O(O(n)*O(n)) ==> O(n^2)
#En realidad resulta en O(n^2) puesto que tiene que recorrer los n elementos y a su vez, revisa si son iguales target y prueba en las n iteraciones.
```
